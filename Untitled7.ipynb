```python
from transformers import OwlViTProcessor, OwlViTForObjectDetection
from PIL import Image
import torch
import numpy as np
import matplotlib.pyplot as plt

model = OwlViTForObjectDetection.from_pretrained("google/owlvit-base-patch32")
processor = OwlViTProcessor.from_pretrained("google/owlvit-base-patch32")
```

```python
image_path = "/content/image1.jpeg"
image = Image.open(image_path).convert("RGB")

text_queries = ["red colored box", "yellow colored box", "green colored box", "blue colored box", "Pink colored Box", "Orange Colored Box"]
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
inputs = processor(text=text_queries, images=image, return_tensors="pt").to(device)
model = model.to(device)
model.eval()

with torch.no_grad():
    outputs = model(**inputs)
logits = outputs.logits
boxes = outputs.pred_boxes
scores = torch.sigmoid(logits).squeeze(0).max(dim=1).values.cpu().numpy()
labels = torch.argmax(logits, dim=2).squeeze(0).cpu().numpy()
boxes = boxes.squeeze(0).cpu().numpy()

width, height = image.size
absolute_boxes = []
for box in boxes:
    cx, cy, w, h = box
    x1 = (cx - w / 2) * width
    y1 = (cy - h / 2) * height
    x2 = (cx + w / 2) * width
    y2 = (cy + h / 2) * height
    absolute_boxes.append([x1, y1, x2, y2])

detections = list(zip(scores, labels, absolute_boxes))
confidence_threshold = 0.01
filtered_detections = [det for det in detections if det[0] >= confidence_threshold]
filtered_detections = sorted(filtered_detections, key=lambda x: x[0], reverse=True)[:5]
```

```python
def map_environment(detections, text_queries, width, height):
    environment_map = []
    print("\n--- Environment Map ---")

    for i, (score, label, box) in enumerate(detections):
        x1, y1, x2, y2 = box
        block_width = x2 - x1
        block_height = y2 - y1

        center_x = (x1 + x2) / 2
        center_y = (y1 + y2) / 2

        environment_map.append({
            "Block Color": text_queries[label],
            "Score": score,
            "Top-Left": (x1, y1),
            "Bottom-Right": (x2, y2),
            "Center": (center_x, center_y),
            "Width": block_width,
            "Height": block_height
        })

        print(f"Block {i+1}:")
        print(f"  Color: {text_queries[label]}")
        print(f"  Score: {score:.2f}")
        print(f"  Position: Top-Left({x1:.2f}, {y1:.2f}), Bottom-Right({x2:.2f}, {y2:.2f})")
        print(f"  Center: ({center_x:.2f}, {center_y:.2f})")
        print(f"  Width: {block_width:.2f}, Height: {block_height:.2f}")
        print()

    environment_map = sorted(environment_map, key=lambda k: (k['Top-Left'][1], k['Top-Left'][0]))

    return environment_map

environment_map = map_environment(filtered_detections, text_queries, width, height)

def plot_environment_map(image, environment_map):
    plt.figure(figsize=(10, 10))
    plt.imshow(image)
    ax = plt.gca()

    for block in environment_map:
        x1, y1 = block['Top-Left']
        x2, y2 = block['Bottom-Right']
        ax.add_patch(plt.Rectangle((x1, y1), x2-x1, y2-y1, edgecolor='red', facecolor='none', linewidth=2))
        ax.text(
            x1, y1, f"{block['Block Color']}: {block['Score']:.2f}",
            color='red', fontsize=12, bbox=dict(facecolor='white', alpha=0.8, edgecolor='red')
        )

    plt.axis('off')
    plt.show()
plot_environment_map(image, environment_map)
```
